# Kmeans clustering

```python
np.random.seed(42)

from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=3, init='random',
                        n_init=1, max_iter=10)
C = kmeans.fit_predict(X) #fa il fit sui dati e fa il predict

colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
```

# Hierarchical clustering

```python
X, _ = datasets.make_blobs(n_samples=50, random_state=8, centers=4)
# X, _ = datasets.make_circles(n_samples=200, factor=.1, noise=.05)

scatter(X[:, 0], X[:, 1]);
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='average', metric='euclidean')#il method serve per dicidere quale distanza tra i punti prendere
figure(figsize=(15,5))
P = dendrogram(Z); #gerarchia dell'albero gerarchico
```

```python
T = fcluster(Z, 4, 'maxclust')

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[T-1]);
```

### Hierarchical Clustering in Scikit-Learn:

```python
from sklearn import cluster

model = cluster.AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')

clusters = model.fit_predict(X)

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[clusters-1]);

#PEr avere una overview del clustering methods andare su scikit-learn.org/stable/modules/clustering.html#clustering
```

# Excercise

Test the different clustering approaches with a "circles" dataset.

```python
from sklearn import datasets

X, _ = datasets.make_circles(n_samples=200, factor=.5, noise=.05)

plot(X[:, 0], X[:, 1], 'bx')
```

```python
#KMeans
np.random.seed(42)

from sklearn import cluster

kmeans = cluster.KMeans(n_clusters=2, init='random',
                        n_init=1, max_iter=10)
C = kmeans.fit_predict(X) #fa il fit sui dati e fa il predict

colors = array(['b', 'r', 'g', 'k','m','c'])
scatter(X[:, 0], X[:, 1], marker="x", c=colors[C])

centers = kmeans.cluster_centers_
scatter(centers[:, 0], centers[:, 1], s=100, c=colors)
```

```python
#Hierarchical Clustering
from sklearn import cluster

model = cluster.AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')

clusters = model.fit_predict(X)

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k','c','y'])
scatter(X[:, 0], X[:, 1], color=colors[clusters-1]);
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

Z = linkage(X, method='single', metric='euclidean')
figure(figsize=(15,5))
P = dendrogram(Z); 
```

```python
T = fcluster(Z, 2, 'maxclust')

figure(figsize=(5,5))
colors = array(['b', 'r', 'g', 'k'])
scatter(X[:, 0], X[:, 1], color=colors[T-1]);
```
